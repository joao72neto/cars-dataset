```python
vendas = pd.read_csv("../car_prices_1p_2014.csv", usecols=["seller", "year", "make", "model", "trim", "body", "transmission", "vin", 
                                                           "condition", "odometer", "color", "interior", "mmr", "sellingprice", "state",
                                                            "diaSemana", "data", "fuso_horario", "zona"])
```

```python
carros = pd.read_csv("../car_prices_1p_2014.csv", usecols=["year", "make", "model", "trim", "body", "transmission", "vin", "condition", 
                                                           "odometer", "color", "interior"])
```

```python
vendedores = pd.read_csv("../tabelas/vendedores.csv")
```

## Trabalhando na tabela carros para a relação com vendas

```python
carros = carros.drop_duplicates()
```

```python
carros["car_id"] = range(1, len(carros) + 1)
```

```python
len(carros)
```

### Associando a tabela de carros com os vendas

```python
#Mapeando transmissão
vendas = vendas.merge(carros, on=["year", "make", "model", "trim", "body", "transmission", "vin", "condition", 
                                "odometer", "color", "interior"], how="left")
```

```python
vendas = vendas[["seller", "car_id", "mmr", "sellingprice", "state", "diaSemana", "data", "fuso_horario", "zona"]].copy()
```

```python
vendas
```

### Associando a tabela de vendas com os vendedores

```python
#Mapeando marcas
vnd_mapa = dict(zip(vendedores["seller"], vendedores["vdr_id"]))
```

```python
vendas["vnd_vdr_id"] = vendas["seller"].map(vnd_mapa)
```

```python
vendas = vendas[["vnd_vdr_id", "car_id", "mmr", "sellingprice", "state", "diaSemana", "data", "fuso_horario", "zona"]]
```

#### Adicionando id à tabela carros

```python
vendas["vnd_id"] = range(1, len(vendas) + 1)
```

```python
vendas = vendas[["vnd_id", "vnd_vdr_id", "car_id", "mmr", "sellingprice", "state", "diaSemana", "data", "fuso_horario", "zona"]]
```

```python
vendas
```

### Escrevendo os dados no CSV

```python
vendas.to_csv("../tabelas/vendas.csv", index=False)
```
